```python
class Exercise1():   # (r-1 + c-1)! / (r-1)!(c-1)!

    def __init__(self):
        self.rows = int(input("Enter the number of rows: "))
        self.cols = int(input("Enter the number of columns: "))

    def find_paths(self):
        #update the grid with the number of paths to reach each cell
        grid = [[0 for i in range(self.cols)] for j in range(self.rows)]
        
        for i in range(0, self.rows):
            grid[i][0] = 1
        for j in range(0, self.cols):
            grid[0][j] = 1

        for r in range(1, self.rows):
            for c in range(1, self.cols):
                grid[r][c] = grid[r-1][c] + grid[r][c-1]
        
        return grid[self.rows - 1][self.cols - 1]


ex1 = Exercise1()
print(ex1.find_paths())
```

```python
class Exercise2():
    
    def __init__(self, costs):
        self.cost = costs

    def min_cost(self):
        for i in range(2, len(self.cost)):
            # update the costs array to reflect the minimum cost to reach each step
            self.cost[i] = min(self.cost[i-1], self.cost[i-2]) + self.cost[i]
        return min(self.cost[-1], self.cost[-2])

ex2 = Exercise2([1, 100, 1, 1, 1, 100, 1, 1, 100, 1])
print(ex2.min_cost())
```

```python
class Exercise3():

    def __init__(self, arr):
        self.arr = arr

    def array_split(self): 
        max_index = len(self.arr) - 1
        min_index = 0
        while min_index < max_index:
            current_index = min_index + ((max_index - min_index) // 2)
            if (self.arr[current_index - 1] < self.arr[current_index]) and (self.arr[current_index] < self.arr[current_index + 1]):
                # We are in the increasing section
                min_index = current_index 
            elif (self.arr[current_index - 1] > self.arr[current_index]) and (self.arr[current_index] > self.arr[current_index + 1]):
                # We are in the decreasing section
                max_index = current_index
            else:
                break
        self.ascending = self.arr[0:current_index].copy()
        self.descending = self.arr[current_index:].copy()
        return self.arr[0:current_index].copy(), self.arr[current_index:].copy()
    
    def binary_search(self, target, array, asc = True):
        max_index = len(array) - 1
        min_index = 0
        while min_index <= max_index:
            current_index = min_index + ((max_index - min_index) // 2)
            if (target < array[current_index]):
                if asc:
                    # target on the left half
                    max_index = current_index - 1
                else:
                    # target on the right half
                    min_index = current_index + 1 
            elif (target > array[current_index]):
                if asc:
                    # target on the right half
                    min_index = current_index + 1
                else:
                    # target on the left half
                    max_index = current_index - 1
            else:
                return True
        return False
    
    def find_target(self, target):
        ascending, descending = self.array_split()
        return self.binary_search(target, ascending) or self.binary_search(target, descending, asc = False)

ex3 = Exercise3([2,4,6,8,10,12,11,9,7,5,3])
print(ex3.find_target(9))

## done
```

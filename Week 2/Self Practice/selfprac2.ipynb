### Exercise 1: Circular Lists

```python
class Node():
    def __init__(self, item, next = None, prev = None):
        self.item = item
        self.next = next
        self.prev = prev

class SinglyCircularList():

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def isEmpty(self):
        return self.size == 0

    def length(self):
        return self.size

    def append(self, item):
        if self.head == None:
            self.head = Node(item)
            self.tail = self.head
        else:
            self.tail.next = Node(item, next = self.head)
            self.tail = self.tail.next
        self.size += 1

    def prepend(self, item):
        if self.head == None:
            self.head = Node(item)
            self.tail = self.head
        else:
            self.head = Node(item, next = self.head)
            self.tail.next = self.head
        self.size += 1

    # 0 based index
    def delete(self, pos):
        if pos + 1 > self.size or pos < 0:
            raise IndexError("Index out of range")
        if self.isEmpty():
            raise IndexError("List is empty")
        
        if pos == 0:
            item = self.head.item
            self.head = self.head.next
            self.tail.next = self.head
            self.size -= 1
            return item
        
        current = self.head
        for _ in range(pos - 1):
            current = current.next
        item = current.next.item
        current.next = current.next.next
        self.size -= 1
        return item
        
    def access(pos):
        for i in range(pos):
            current = current.next
        return current.item
    
    def __str__(self):
        if self.isEmpty():
            return "[]"
        
        current = self.head
        list = [current.item]
        for _ in range(self.size - 1):
            current = current.next
            list.append(current.item)
        return str(list)
```

```python
list = SinglyCircularList()
list.append(1)
list.append(2)
print(list)
assert list.head.item == 1
assert list.tail.item == 2
assert list.head.next.item == 2
assert list.head.next == list.tail
assert list.tail.next == list.head
list.append(3)
print(list)
list.prepend(0)
print(list)
assert list.head.item == 0
assert list.tail.item == 3
assert list.tail.next == list.head
assert list.size == 4
#new = SinglyCircularList()
print(list.delete(3))
print(list)
print(list.delete(0))
print(list)
assert list.delete(1) == 2
l2 = SinglyCircularList()
try:
    l2.delete(0)
except IndexError as e:
    print("caught")
l2.prepend(10)
print(l2)
assert l2.head.item == 10
assert l2.tail.item == 10
try:
    l2.delete(-3)
except IndexError as e:
    print("caught")
```

```python
class DoublyCircularList():
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def isEmpty(self):
        return self.size == 0
    
    def length(self):
        return self.size
    
    def append(self, item):
        if self.head == None:
            self.head = Node(item)
            self.tail = self.head
            self.head.next = self.head
            self.head.prev = self.head
        else:
            newNode = Node(item, next = self.head, prev = self.tail)
            self.tail.next = newNode
            self.head.prev = newNode
            self.tail = newNode
        self.size += 1

    def prepend(self, item):
        if self.head == None:
            self.head = Node(item)
            self.tail = self.head
            self.head.next = self.head
            self.head.prev = self.head
        else:
            self.head = Node(item, next = self.head, prev = self.tail)
            self.head.next.prev = self.head
            self.tail.next = self.head
            self.head.prev = self.tail
        self.size += 1

    # 0 based index
    def delete(self, pos):
        if pos + 1 > self.size or pos < 0:
            raise IndexError("Index out of range")
        if self.isEmpty():
            raise IndexError("List is empty")
        
        if pos == 0:
            item = self.head.item
            self.head = self.head.next
            self.head.prev = self.tail
            self.tail.next = self.head
            self.size -= 1
            return item
        
        current = self.head
        for _ in range(pos - 1):
            current = current.next
        item = current.next.item
        if current.next == self.tail:
            self.tail = current
        current.next = current.next.next
        current.next.prev = current
        self.size -= 1
        return item
        
    def access(self, pos):
        if pos < 0 or pos >= self.size:
            raise IndexError("Index out of range")
        current = self.head
        for _ in range(pos):
            current = current.next
        return current.item
    
    def __str__(self):
        if self.isEmpty():
            return "[]"
        
        current = self.head
        list = [current.item]
        for _ in range(self.size - 1):
            current = current.next
            list.append(current.item)
        return str(list)
```

```python
double_list = DoublyCircularList()
double_list.append(1)
double_list.append(2)
print(double_list)

assert double_list.head.item == 1
assert double_list.tail.item == 2
assert double_list.head.next.item == 2
assert double_list.head.next == double_list.tail
assert double_list.tail.next == double_list.head
assert double_list.head.prev == double_list.tail
assert double_list.tail.prev == double_list.head

double_list.append(3)
print(double_list)
double_list.prepend(0)
print(double_list)

assert double_list.access(2) == 2
assert double_list.head.item == 0
assert double_list.tail.item == 3
assert double_list.tail.next == double_list.head
assert double_list.head.prev == double_list.tail
assert double_list.size == 4
#new = SinglyCircularList()
print(double_list.delete(3))
print(double_list)
assert double_list.tail.item == 2
assert double_list.tail.prev.item == 1
assert double_list.tail.next.item == 0
assert double_list.tail.next == double_list.head
assert double_list.head.prev == double_list.tail
print(double_list.delete(0))
print(double_list)
assert double_list.tail.item == 2
assert double_list.tail.prev.item == 1
assert double_list.tail.next.item == 1
assert double_list.tail.next == double_list.head
assert double_list.head.prev == double_list.tail

assert double_list.delete(1) == 2
assert double_list.head == double_list.tail
assert double_list.head.next == double_list.head
assert double_list.head.prev == double_list.head
assert double_list.tail.next == double_list.tail
assert double_list.tail.prev == double_list.tail

l2 = DoublyCircularList()
try:
    l2.delete(0)
except IndexError as e:
    print("caught")
l2.prepend(10)
print(l2)
assert l2.head.item == 10
assert l2.tail.item == 10
assert l2.head == l2.tail
assert l2.head.next == l2.head
assert l2.head.prev == l2.head
assert l2.tail.next == l2.tail
assert l2.tail.prev == l2.tail
try:
    l2.delete(-3)
except IndexError as e:
    print("caught")
```

### Exercise 2: Deque

```python
from typing import Any

class Deque():
    def __init__(self):
        self.items = []

    def isEmpty(self) -> bool:
        return len(self.items) == 0

    def addFirst(self, item: Any) -> None:
        self.items.insert(0, item)

    def addLast(self, item: Any) -> None:
        self.items.append(item)

    def removeFirst(self) -> Any:
        if self.isEmpty():
            raise IndexError("Deque is empty")
        return self.items.pop(0)

    def removeLast(self) -> Any:
        if self.isEmpty():
            raise IndexError("Deque is empty")
        return self.items.pop()
```

```python
def testDeque():
    deque = Deque()
    assert deque.isEmpty() == True
    deque.addFirst(1)
    deque.addLast(2)
    print(deque.items)
    assert deque.isEmpty() == False
    assert deque.removeFirst() == 1
    print(deque.items)
    assert deque.removeLast() == 2
    print(deque.items)
    assert deque.isEmpty() == True
    try:
        deque.removeFirst()
    except IndexError as e:
        print("caught")
    try:
        deque.removeLast()
    except IndexError as e:
        print("caught")
    deque.addFirst(1)
    deque.addFirst(0)
    print(deque.items)
    assert deque.removeLast() == 1
    assert deque.removeLast() == 0
testDeque()
```

### Exercise 3: Palindrome

```python
def check_palindrome(string: str) -> bool:
    if string is None:
        raise ValueError("Input string cannot be None")
    i = 0
    j = len(string) - 1
    string = string.lower()
    while j > i:
        if string[i] != string[j]:
            return False
        i += 1
        j -= 1
    return True

def test_palindrome_checker():
    assert check_palindrome("racecar") == True
    assert check_palindrome("tea") == False
    assert check_palindrome("abba") == True
    assert check_palindrome("AbbBa") == True
    assert check_palindrome("radar") == True
    assert check_palindrome("Topspot") == True
    assert check_palindrome("brit") == False
    try:
        check_palindrome(None)
    except ValueError as e:
        assert str(e) == "Input string cannot be None"
test_palindrome_checker()
```

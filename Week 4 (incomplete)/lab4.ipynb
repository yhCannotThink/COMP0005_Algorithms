```python
class LabWeek4():

    def __init__(self):
        self.flip_log = []

    def flip(self, list, index):
        list = list[index::-1] + list[index+1:]
        #print(list)
        return list

    def find_connected(self, list):
        asc = False
        desc = False
        breakpoint = []
        for i in range(len(list) - 1):
            if list[i] + 1 == list[i + 1]:
                asc = True
            elif list[i] - 1 == list[i + 1]:
                desc = True
            else:
                breakpoint.append(i)
        return breakpoint
    
    def find_largest(self, list, abs_largest):
        largest = 0
        for i in range(len(list) - 1):
            if list[i] > largest and list[i] < abs_largest:
                largest = list[i]
                index = i
        return largest, index

    def bad_pancake_sort(self, list):
        abs_largest = 100000
        sorted = len(list) - 1
        self.flip_log = []
        while sorted != 0:
            largest, largest_index = self.find_largest(list, abs_largest)
            if largest_index != 0:
                list = self.flip(list, largest_index)
                self.flip_log.append(largest_index)
            if list.index(largest) != sorted:
                list = self.flip(list, sorted)
                self.flip_log.append(sorted)
            abs_largest = largest
            sorted -= 1
        print("sorted list", list)
        print("indices flipped", self.flip_log)

    def reversal(self, list, start, end):
        list = list[:start - 1] + list[end-1:start-2:-1] + list[end:]
        print(list)
        return list

    def pancake_sort_rev(self, list):
        breakpoint = self.find_connected(list)
        print(breakpoint)
        for i in range(len(breakpoint)+1):
            list_clone = list.copy()
            list_clone = self.reversal(list_clone, breakpoint[i], breakpoint[i+1]+1)
            breakpoint = self.find_connected(list)
            print(breakpoint)

q = LabWeek4()
q.bad_pancake_sort([3,2,4,1])

q.bad_pancake_sort([3,5,7,9,4,8,1,2,6])

q.pancake_sort_rev([1, 8, 9, 3, 2, 7, 6, 5, 4, 10])
```

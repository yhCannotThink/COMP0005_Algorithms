```python
from time import time
from timeit import default_timer as timer

class Exercise1():

    def linear_search(self, target, list):
        start_time = timer()
        calls = 0
        for index in range(len(list)):
            calls += 1
            if target == list[index]:
                end_time = timer()
                return index, calls, (end_time - start_time)
        end_time = timer()
        return -1, calls, (end_time - start_time)

    def binary_search(self, target, list, low_index, hi_index, start_time, calls):
        if low_index > hi_index:
            end_time = timer()
            return -1, calls, (end_time - start_time)
        index = low_index + (hi_index - low_index) // 2
        if list[index] == target:
            end_time = timer()
            return index, calls, (end_time - start_time)
        elif list[index] < target:
            return self.binary_search(target, list, index + 1, hi_index, start_time, calls + 1)
        else:
            return self.binary_search(target, list, low_index, index - 1, start_time, calls + 1)

    def trinary_search(self, target, list, low_index, hi_index, start_time, calls):
        if low_index > hi_index:
            end_time = timer()
            return -1, calls, (end_time - start_time)
        q1 = (hi_index - low_index) // 3 + low_index
        q3 = (hi_index - low_index) * 2 // 3 + low_index
        if list[q1] == target:
            end_time = timer()
            return q1, calls, (end_time - start_time)
        elif list[q3] == target:
            end_time = timer()
            return q3, calls, (end_time - start_time)
        elif target < list[q1]:
            return self.trinary_search(target, list, low_index, q1 - 1, start_time, calls + 1)
        elif target > list[q3]:
            return self.trinary_search(target, list, q3 + 1, hi_index, start_time, calls + 1)
        else:
            return self.trinary_search(target, list, q1 + 1, q3 - 1, start_time, calls + 1)
        
#import random

class Exercise2():

    def gen_list(self, number):
        self.list = [i for i in range(number)]
        #print(self.list)

    def experiment(self):

ex1 = Exercise1()
#print(ex1.linear_search(4, [1,2,3,4,5,7,8,9]))
#print(ex1.binary_search(9, [1,2,3,4,5,7,8,9], 0, 7))
#print(ex1.trinary_search(3, [1,2,3,4,5,7,8,9], 0, 7))

ex2 = Exercise2()
ex2.gen_list(1000)
print(ex1.linear_search(500, ex2.list))
print(ex1.binary_search(500, ex2.list, 0, 999, timer(), 1))
print(ex1.trinary_search(500, ex2.list, 0, 999, timer(), 1))
```

```python
class Exercise5():

    def __init__(self, columns, rows):
        self.row = rows
        self.matrix = {}
        for i in range(columns):
            self.matrix[i] = {}

    def set_matrix(self, matrix_array):
        if len(matrix_array) > len(self.matrix) * self.row:
            raise ValueError("Too many values in matrix")
        counter = 0
        for col in range(len(self.matrix)):
            for row in range(self.row):
                if matrix_array[counter] != 0:
                    self.matrix[col][row] = matrix_array[counter]
                counter += 1

    def add(self, column, row, value):
        self.matrix[column][row] = value

    def print_matrix(self):
        for i in range(len(self.matrix)):
            for j in range(self.row):
                if j not in self.matrix[i]:
                    print("0", end = " ")
                else:
                    print(self.matrix[i][j], end=" ")
            print() 

    def matrix_sum(self, other_matrix):
        if len(self.matrix) != len(other_matrix.matrix) or self.row != other_matrix.row:
            raise ArithmeticError("The matrices are of different dimensions")
        for (k, v) in self.matrix.items():
            for (key, val) in self.matrix[k].items():
                if key not in other_matrix.matrix[k]:
                    other_matrix.matrix[k][key] = val
                else:
                    other_matrix.matrix[k][key] += val
        return other_matrix
    
    def matrix_multiplication(self, other_matrix):
        if self.row != len(other_matrix.matrix):
            raise ValueError("The matrix dimensions are incompatible")
        new_matrix = self.matrix.copy()
        print(new_matrix)
        new_matrix[0][0] = 0
        for i in range(self.row):
            new_matrix[0][0] += self.matrix[0][i] * other_matrix.matrix[i][0]
        return other_matrix
    
#    1 2 3      1 2 3
#    4 5 6  *   4 5 6
#    7 8 9      0 0 0

A = Exercise5(3, 3)
A.set_matrix([1, 2, 3, 0, 5, 6, 7, 8, 9])
#A.print_matrix()
B = Exercise5(3, 3)
B.set_matrix([1,2,3,4,5,6,0,0,0])
#B.print_matrix()
#print()
D = A.matrix_multiplication(B)
C = A.matrix_sum(B)
#C.print_matrix()
```

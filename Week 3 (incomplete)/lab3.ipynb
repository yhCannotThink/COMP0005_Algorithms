```python
class Anagrams():
    def __init__(self, words):
        self.words = words
        self.dict = {}

    def sort_word(self):
        for word in self.words:
            self.dict[word] = sorted(word)

    def group(self):
        self.sort_word()
        self.dict = dict(sorted(self.dict.items(), key=lambda item: item[1]))
        current_list = []
        last_word = self.dict[list(self.dict.keys())[0]]
        for word, sorted_word in self.dict.items():
            if sorted_word == last_word:
                current_list.append(word)
            else:
                last_word = sorted_word
                print(current_list)
                current_list = [word]
        print(current_list)
            
ex1 = Anagrams(["eat", "tea", "part", "ate", "trap", "pass"])
ex1.group()
```

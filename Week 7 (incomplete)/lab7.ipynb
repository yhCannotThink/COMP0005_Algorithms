```python
class DirectedGraph():

    def __init__(self, V):
        self.V = V
        self.adj = []
        for _ in range(0, V):
            self.adj.append(set())

    def addEdge(self, v, w):
        self.adj[v].add(w)

    def adj(self, v):
        return self.adj[v]

ex1 = DirectedGraph(6)
ex1.addEdge(0, 1)
ex1.addEdge(2, 0)
ex1.addEdge(0, 5)
ex1.addEdge(2, 3)
ex1.addEdge(3, 5)
ex1.addEdge(4, 3)
ex1.addEdge(4, 2)
ex1.addEdge(3, 2)
ex1.addEdge(5, 4)
#print(ex1)
print(ex1.adj)
#print(len(ex1.adj[1]))
```
